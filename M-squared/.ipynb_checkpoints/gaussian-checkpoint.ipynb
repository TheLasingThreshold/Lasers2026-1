```python
import cv2, os
import numpy as np
import matplotlib.pyplot as plt
```

```python
filedir = "85-0mA_Raw_ZScan_0-2mm"

for filename in os.listdir(filedir):
    print(filename)
    img = cv2.imread(os.path.join(filedir, filename))

fig, ax = plt.subplots()
im = ax.imshow(img)
ax.set(title=filename)
```
